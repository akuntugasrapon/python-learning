## 1. Apa itu Machine Learning?

Machine Learning adalah cabang dari Artificial Intelligence (AI)
yang memungkinkan komputer belajar dari data tanpa diprogram secara eksplisit.

Machine Learning digunakan dalam:
- Prediksi harga
- Pengenalan wajah
- Sistem rekomendasi
- Analisis data

## 2. Peran Python dalam Machine Learning

Python adalah bahasa pemrograman yang populer untuk Machine Learning karena:
- Sintaks sederhana
- Banyak library (NumPy, Pandas, Scikit-learn)
- Komunitas besar


```python
# Dasar Python

print("Belajar Python untuk Machine Learning")

a = 10
b = 5

print("Penjumlahan:", a + b)
print("Perkalian:", a * b)

data = [1, 2, 3, 4, 5]
print("Data:", data)
print("Rata-rata:", sum(data)/len(data))
```

```python
# Import library untuk Machine Learning
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Contoh Machine Learning sederhana

# Data contoh
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])

# Membuat model
model = LinearRegression()

# Melatih model
model.fit(X, y)

# Prediksi nilai baru
prediksi = model.predict([[6]])

print("Prediksi untuk input 6 adalah:", prediksi[0])
```

```python
# Visualisasi hasil model
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression Example")
plt.show()
```
