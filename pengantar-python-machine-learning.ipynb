# Pengantar Python untuk Machine Learning

## Apa itu Machine Learning?

Machine Learning adalah cabang dari Artificial Intelligence (AI) 
yang memungkinkan komputer belajar dari data tanpa diprogram secara eksplisit.

## Mengapa Python digunakan untuk Machine Learning?

- Sintaks sederhana
- Banyak library seperti NumPy, Pandas, dan Scikit-learn
- Mudah dipahami oleh pemula


```python
# Dasar Python

print("Belajar Python untuk Machine Learning")

a = 10
b = 5

print("Penjumlahan:", a + b)
print("Perkalian:", a * b)

data = [2, 4, 6, 8, 10]
print("Data:", data)
print("Rata-rata:", sum(data)/len(data))
```

```python
# Import library Machine Learning
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Contoh Machine Learning sederhana

X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])

model = LinearRegression()
model.fit(X, y)

prediksi = model.predict([[6]])

print("Prediksi untuk input 6 adalah:", prediksi[0])
```
